### Outlier Treatment

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3-q1
upper = q3+1.5*iqr
lower = q1-1.5*iqr

trimmed_data = df[((df>upper) | (df<lower)).any(axis=1)]
trimmed_data.head()
```

### Data without outliers

```python
trimmed_data_without_outlier = df[~((df>upper) | (df<lower)).any(axis=1)]
trimmed_data_without_outlier.head()
```
